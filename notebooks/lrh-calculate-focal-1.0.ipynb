```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 1
x = np.linspace(0, 1, 100)

# Calculate y values for each function
y1 = np.ones_like(x)  # y = 1
y2 = 2 * x  # y = 2x
y3 = 3 * x**2  # y = 3x^2
y4 = 4 * x**3  # y = 4x^3

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the curves and fill the areas
plt.fill_between(x, y1, alpha=0.3, label='y=1')
plt.fill_between(x, y2, alpha=0.3, label='y=2x')
plt.fill_between(x, y3, alpha=0.3, label='y=3x')
plt.fill_between(x, y4, alpha=0.3, label='y=4x')

# Plot the lines on top
plt.plot(x, y1, 'b-')
plt.plot(x, y2, 'r-')
plt.plot(x, y3, 'g-')
plt.plot(x, y4, 'y-')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparison of 1, 2x, 3x, and 4x')
plt.legend()
plt.grid(True)
plt.show()
```
